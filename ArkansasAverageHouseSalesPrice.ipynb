### **Sorting Data**

```python
# Sorting data
house_prices_sorted_df = house_prices_df.sort_values(by=['State', 'Metro Area', 'RegionName', 'CountyName'])
house_prices_sorted_df.head()
```

### **Filtering Data**

```python
# filtering data
# Arkansas house prices
AR_house_prices_df = house_prices_sorted_df[(house_prices_sorted_df['State'] == 'AR')]
AR_house_prices_df.head()
```

### **Transforming Data**

```python
# drop Size Rank column
AR_house_prices_df.drop(['SizeRank'], axis = 1, inplace = True)
AR_house_prices_df.head()
```

```python
# drop RegionID column
AR_house_prices_df.drop(['RegionID'], axis = 1, inplace = True)
AR_house_prices_df.head()
```

### **Null Values**

```python
# Null Values
AR_house_prices_df.isna()
```

AR_house_prices_df.head()

### **Pivot Tables**

```python
pivot_AR_house_prices_df = pd.pivot_table(AR_house_prices_df, index = ['Metro Area', 'CountyName'], values = '2015-01')
pivot_AR_house_prices_df
```

```python
# pivot bar chart
# Create Numpy Arrays
x = np.array(["Arkadelphia", "Fayetteville", "Fort Smith", "Hot Springs", "Little Rock", "Memphis"])
y = np.array([52500.000000, 141965.384615, 99560.000000, 162066.666667, 123970.967742, 97350.000000])

# Create New Figure
AR_Pivot_Bar_Chart = plt.figure()

# Create Bar Chart
plt.bar(x, y, color = ['darkred', 'gold', 'lime', 'aqua', 'navy', 'darkmagenta'])

# Add Chart Title
plt.title("Average House Sales Price During 2015-01")

# Add Axis Titles
plt.xlabel('Metro Area')
plt.ylabel('Average House Sales Price')

# Add Grid Lines
plt.grid(True)

```

### **Which Arkansas metro area had the highest average house sales price during 2015-01?**

### **Which Arkansas metro during 2015-01 had the lowest average house sales price?**
